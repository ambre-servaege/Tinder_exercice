Le but de l'exercice est d'expliquer à travers les speed dating de 4 minutes, en quoi les gens sont attirés par les autres.
Après le speed dating, on demande aux participants si ils veulent revoir le partenaire ou non et de noter le partenaire selon ces critères :
- attractivité
- sincérité
- intelligence
- fun
- ambition
- intérêts partagés

```python
#Maintenant que le dataset est propre nous allons analyser et comparer les données entre elles
```

```python
# Calculer le nombre d'hommes et de femmes
nombre_hommes = df['gender'].value_counts().get(1, 0)
nombre_femmes = df['gender'].value_counts().get(0, 0)

# Afficher les résultats
print(f"Nombre d'hommes : {nombre_hommes}")
print(f"Nombre de femmes : {nombre_femmes}")
```

```python
#On calcule la moyenne de partenaires recontrés dans une sessions
```

```python
moyenne_rencontre_session = df['round'].mean().round(1)
print(moyenne_rencontre_session)
```

```python
# Calculer le nombre de matchs et de non-matchs
nombre_de_matchs = df['match'].value_counts().get(1, 0)
nombre_de_non_matchs = df['match'].value_counts().get(0, 0)

# Afficher les résultats
print("Nombre de matchs : ", nombre_de_matchs)
print("Nombre de non-matchs : ", nombre_de_non_matchs)
```

```python
# Calculer le nombre de personnes étant de la même ethnie
idem_ethnie = df['samerace'].value_counts().get(1, 0)
diff_ethnie = df['samerace'].value_counts().get(0, 0)

# Afficher les résultats
print("Personne de la même ethnie : ", idem_ethnie)
print("Personne d'une ethnie différente : ", diff_ethnie)
```
